## <font color='violet'>1- the concept of expedcted return  of the rewards at a given time  </font>

mathematically, the return is the sum of future rewards:



the agent objective is to maximize the expected return, so the expected return is what drives the agent to make the decisions that it makes. 

Here, the agent environment interaction is naturally breaks up into **episodes**, and each episode ends by giving this agent a positive or negitive reward, then the environment is resetted to some standard starting state or to a random sample from a normal distribution of possible starting states, the next episode begins indpendately from how the previous episode ended. this kind of tasks is called **episodic tasks**. 

but in real life, we have a kind of taks performed continuously without stopping, this is called **continuing tasks**, the above equation can't fit this type of tasks as the final reward will setted at time infinity and therefore the return itself can be infinite. 


### and now the discounted return plays the role

instead of maximizing the return of the rewards, the agent will try to maximize the **discounted** return of the rewards by multiplying each reward in a weighted parameter called **discount rate (gamma)**.


**Notes** 

- the discounted rate is a number between 0 and 1.

- the discount rate will determine the present value for future rewards

##   <font color='violet'>2- optimal policies and optimal value funciton </font>

we want to find the policy that will yield the highest return possible if the agent followed this policy. so we say that:

a policy **pi** is better than or equel to policy **pi prime** if the **value of state S under policy pi** >= **the value of state S under policy policy pi prime ** for all the states



we have two types of optimal funcitons:

###    <font color='black'>2-1 optimal state-value function( optimal v funciton)  </font>

this value returns the maximum value of that state under any given policy, so we're trying here to have the largest expected return achievble by any policy **pi** for each possible state.



###   <font color='black'> 2-2 optimal action-value function( optimal Q funciton)  </font>

this value returns the maximum state-action value of that state under any given policy, so we're trying here to have the  largest expected return achievble by any policy **pi** for each possible state-action pair.

###  <font color='black'> 2-3  the bellman optimality equation for q </font>


with optimal q function we're trying to find the optimal policy. for each state, we want to make the action that has the highest q value for the current state.


it states that: the expected return from starting in state **s** and taking an action **a** then follow a policy **pi** is going to be **the expected reward we get from that action a in state s + the maximum expected discounted return that can be achievable by any possible next state-action pair (s prime, a prime)**



**Note**: 

**s prime** here the best possible state for next action **a prime** can be taken at time t+1. 

 


##   <font color='violet'>3 - exploitation and exploration </font>

we said that we need a balance between exploitation and exploration, to get that balance we will use something called 
### **epsilon greedy strategy**, 
with this strategy we will define **epsilon rate** with initial value **1** , this exploration rate is the probability that our agent will explore the environment rather than exploit it. with **epsilon=1** it's a hundered percent that our agent will start to explore the environmnet. at the start of each new episode , the agent will decay the epsilon value by some rate as it getting to know the environment more and more, after that, the agent tends to be greedy to exploit the environment once it had the oppurtunity to explore it. 

to determine whether the agent will tend to exploit or explore the environment, we will generate a ramdom number between 0 and 1:

--> if it's greater than epsilon, it means that the agent will tend to exploit the environment (it will choose the value of the highest Q for its current state)

--> otherwise, the agent will tend to explore the environment (ranomly choosing its action and explore what will happend in the environment)
